```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#connect to URL

year = 2020
url = requests.get('https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'.format(year)).text
```

```python
#read it as lxml
soup = BeautifulSoup(url,'lxml')
```

```python
#either inspect the element or find it yourself here
print(soup.prettify())
```

```python
soup.findAll('tr', limit=2)
headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]

#Basketball Reference for the analysis
headers = headers[1:]
headers
```

```python
# avoid the first header row
rows = soup.findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
```

```python
#set up as pandas dataframe
df_2020 = pd.DataFrame(player_stats, columns = headers)
```

```python
#add the season back into the data
df_2020.insert(1,'Season','2019-20')
df_2020.head(10)
df_2020 = df_2020.dropna()
df_2020
```
