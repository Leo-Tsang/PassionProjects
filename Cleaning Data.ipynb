```python
import pandas as pd
```

```python
#finding the conference of the teams

df = pd.read_csv('nba_2015-2018')
df.head(10)
```

# Insert Conference

```python
East = set(['MIL', 'BOS', 'MIA', 'PHI', 'TOR', 'IND', 'BRK', 'ORL', 'DET', 'CHO','CLE', 'CHI', 'WAS', 'ATL', 'NYK'])
Team = set(df['Tm'])
West = Team - East
```

```python
#add the conference as a list to the dataframe
conf_list = []
for i in list(df['Tm']):
    if i in East:
        conf_list.append('East')
    if i in West:
        conf_list.append('West')

#insert this data right beside the team column
df.insert(4,'Conf', conf_list)
```

```python
#All star posistions are mixed between F and G
pos_new = []
for i in list(df['Pos']):
    if i == 'G' or i == 'SG':
        pos_new.append('G')
    else:
        pos_new.append('F')
        
#insert this data after pos column
df.insert(3, 'Pos2', pos_new)
```
